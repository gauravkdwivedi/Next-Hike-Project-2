```python
dataset3 = pd.read_csv('C:/Users/gaura/Downloads/dataset_3.csv')
dataset3
```

```python
concatenated_dataset2 = pd.concat([merged_dataset, dataset3])
concatenated_dataset2
```

```python
df2 = pd.DataFrame(concatenated_dataset2)
null_values = df2.isnull()
null_counts = null_values.sum()
print(null_counts)
```

```python
q1 = concatenated_dataset2[['temp','atemp', 'hum', 'windspeed']].quantile(.25)
q3 = concatenated_dataset2[['temp','atemp', 'hum', 'windspeed']].quantile(.75)
```

```python
def find_outliers (df2, cols):
    q1 = df2[cols].quantile(.25)
    q3 = df2[cols].quantile(.75)
    iqr = q3 - q1
    min_r = q1 -1.5*iqr
    max_r = q3 + 1.5*iqr
    outliers_indices = df2.index[(df2[cols] < min_r) | (df2[cols] > max_r)]
    return outliers_indices
```

```python
find_outliers(df2, 'temp')
```

```python
find_outliers(df2, 'atemp')
```

```python
find_outliers(df2, 'hum')
```

```python
find_outliers(df2, 'windspeed')
```
